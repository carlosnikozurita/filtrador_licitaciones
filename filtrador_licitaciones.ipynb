# FILTRO BASES DE DATOS LICITACIONES

```python
import pandas as pd
from IPython.display import display

# Reemplaza 'ruta/del/archivo.csv' con la ruta real de tu archivo CSV

try:
    df = pd.read_csv(r'C:\Users\i732gb 512x2 rtx2070\Downloads\Licitacion_Publicada3\Licitacion_Publicada.csv', header=2)
except pd.errors.ParserError as e:
    print("Error al analizar el archivo CSV:", e)

# Ajustar la configuración de visualización para mostrar más filas y columnas
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Palabras clave adicionales
palabras_clave_adicionales = ['áreas verdes', 'bosque', 'nativo', 'ambiental',
                              'paisajismo', 'flora', 'fauna', 'vegetación',
                              'parque', 'jardín', 'conservación', 'ecología',
                              'sostenibilidad', 'reserva natural', 'biodiversidad',
                              'arboricultura', 'restauración',
                              'ecosistema', 'jardinería','árbol', 'naturaleza',
                              'parque urbano', 'manejo forestal', 'ecosistema',
                              'medio ambiente', 'conservación del suelo', 'corte'
                             ]

# Filtrar filas que contienen "keywords" en las columnas Textbox37 o Textbox38
filtro_keywords = (
    df['Textbox37'].str.contains('|'.join(palabras_clave_adicionales), case=False) |
    df['Textbox38'].str.contains('|'.join(palabras_clave_adicionales), case=False)
)    

# Filtrar filas que tienen la región compradora como "Región de los Lagos" o "Región de Los Ríos"
filtro_regiones = df['citName'].str.contains('Región de los Lagos|Región de Los Ríos', case=False)

# Aplicar ambos filtros al DataFrame original
df_filtrado = df[filtro_keywords & filtro_regiones]

# Imprimir el DataFrame resultante en formato tabular
display(df_filtrado)

# Restaurar la configuración de visualización predeterminada
pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')

# Exportar el DataFrame filtrado a un archivo Excel
df_filtrado.to_excel('resultado_filtrado.xlsx', index=False)
```
